```python
from IPython.display import clear_output

dico = {1:' ',2:' ',3:' ',
        4:' ',5:' ',6:' ',
        7:' ',8:' ',9:' '}

def theboard(dico):
    clear_output()
    print(" | ",dico[1]," | ",dico[2]," | ",dico[3]," | ")
    print(" ------+-----+------")
    print(" | ",dico[4]," | ",dico[5]," | ",dico[6]," | ")
    print(" ------+-----+------")
    print(" | ",dico[7]," | ",dico[8]," | ",dico[9]," | ")   

def gameover(theboard):
    return(theboard(dico))
#THIS ONE IS NOT REALLY NECESSARY 

def replay(game):
    for n in range(1,10):
        dico[n] = ' '
    game()

#THE GAME STARTS HERE____________________________________________________________________________________________________
def game():
    
    theboard(dico)
    print("\nX - It's your turn")
    x = int(input())
    dico[x] = 'X'
    gameover(theboard)
    i = 0
    roll = True
    tie = [x]
    print("\n0 - It's your turn")
    
 #WHILE LOOP STARTS HERE___________________________________________________________________________________________________    
    while roll:
        if len(tie) < 9:
            if not(dico[1] == dico[2] == dico[3] == dico[4] == dico[5] == dico[6] == dico[7] == dico[8] == dico[9] == ' '):
                x = int(input())

                if dico[x] == ' ':
                    if i%2 != 0:
                        dico[x] = 'X'
                        tie.append('0')
                    else:
                        dico[x] = '0'
                        tie.append('X')

 #Verification Horizontale------------------------------------------------------------------------------------------------
                    if (dico[1] != ' ' and dico[2] != ' ' and dico[3] != ' ' and dico[1] == dico[2] == dico[3]):
                        roll = False
                        gameover(theboard)
                    elif (dico[4] != ' ' and dico[5] != ' ' and dico[6] != ' ' and dico[4] == dico[5] == dico[6]):
                        roll = False
                        gameover(theboard)
                    elif (dico[7] != ' ' and dico[8] != ' ' and dico[9] != ' ' and dico[7] == dico[8] == dico[9]):
                        roll = False
                        gameover(theboard)

 #Verification Verticale--------------------------------------------------------------------------------------------------
                    elif (dico[1] != ' ' and dico[4] != ' ' and dico[7] != ' ' and dico[1] == dico[4] == dico[7]):
                        roll = False
                        gameover(theboard)
                    elif (dico[2] != ' ' and dico[5] != ' ' and dico[8] != ' ' and dico[2] == dico[5] == dico[8]):
                        roll = False
                        gameover(theboard)
                    elif (dico[3] != ' ' and dico[6] != ' ' and dico[9] != ' ' and dico[3] == dico[6] == dico[9]):
                        roll = False
                        gameover(theboard) 

 #Verification Diagonale-------------------------------------------------------------------------------------------------
                    elif (dico[1] != ' ' and dico[5] != ' ' and dico[9] != ' ' and dico[1] == dico[5] == dico[9]):
                        roll = False
                        gameover(theboard)
                    elif (dico[3] != ' ' and dico[5] != ' ' and dico[7] != ' ' and dico[3] == dico[5] == dico[7]):
                        roll = False
                        gameover(theboard)

                    else:
                        i += 1
                        #tie.append(dico[x])
                        theboard(dico)
                        print(f"\n{tie[-1]} - It's your turn")
                else:
                    print('ERROR')
        else:
            roll = False
    else:
        if roll == False and len(tie) == 9:
            print('TIE GAME \n')
        else:
            print('YOU WON \n')
            
        print('Do you want to replay the game?')
        x = input()
        if x.lower() == "yes":
            replay(game)
        else:
            print('Thanks for playing!')
```
